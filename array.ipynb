Q.1 maximum and minimum element in an array

```python
#minimum maximum element using comparison
a =[3,5,4,1,9]
minimum_element = 3
maximum_element = -1
for i in a:
     if i < minimum_element:
          minimum_element = i
for i in a:
     if i > maximum_element:
          maximum_element = i
print("the maximum element is :",maximum_element)
print("The minimum element is :",minimum_element)
# time complexity is O(n)
```

```python
# maximum and minimum element using sorting
a = [3, 5, 4, 1, 9]
a.sort()
print("The minimum element is :", a[0])
print("The maximum element is :", a[-1])
# time complexity is O(nlogn)
```

maximum subarray

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
max_sum = -1
current_sum = 0
for i in nums:
    current_sum += i
    if current_sum > max_sum:
        max_sum = current_sum
    if current_sum < 0:
        current_sum = 0
print("The maximum subarray sum is :", max_sum)
```
